```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

get_ipython().ast_node_interactivity = 'all'
pd.options.display.max_columns = None

df = pd.read_csv('source_files/mm_challenge.csv', index_col=False)

df.head()
```

```python
# copy only the needed fields

df_reduced = df[['mnppid', 'edbirthd', 'edsex']].copy()
df_reduced.head()
df_reduced.shape
```

```python
# check for duplicates

df_reduced[df_reduced.mnppid.duplicated(keep=False)]
```

```python
# check sex column for unforseeen/unexpected values

df_reduced.edsex.unique()
(df_reduced['edsex'].values == ' ').sum()
df_reduced['edsex'].isnull().sum()
(df_reduced['edbirthd'].values == ' ').sum()
```

```python
# standard birth date column

df_reduced.edbirthd.head()
df_reduced['edbirthd']= df_reduced.edbirthd.replace(r'^\s*$', np.nan, regex=True)
df_reduced['edbirthd'] = pd.to_datetime(df_reduced.edbirthd)
df_reduced.edbirthd.head()
```

```python
# transform birth date to seperate year, month, day columns
df_reduced['year'] = df_reduced['edbirthd'].dt.year
df_reduced['month'] = df_reduced['edbirthd'].dt.month
df_reduced['day'] = df_reduced['edbirthd'].dt.day
df_reduced.head()
df_reduced.iloc[6266]
df_reduced = df_reduced.drop('edbirthd', axis=1)
df_reduced.head()
pd.isnull(df_reduced).any(1).to_numpy().nonzero()
df_reduced.day[6266]
```

```python
# create person table
person_df = pd.DataFrame(columns=['person_id','person_source_value', 'gender_concept_id', 
                                  'gender_source_value', 'year_of_birth',
                                  'month_of_birth', 'day_of_birth'])
person_df
```

```python
# Map source columns to OMOP person table columns
person_df['person_id'] = df_reduced.index
person_df['person_source_value'] = df_reduced['mnppid']
person_df['gender_concept_id'] = df_reduced['edsex'].map({'Male': 8507, 'Female': 8532, 'np.nan': 0})
person_df['gender_source_value'] = df_reduced['edsex']
person_df['year_of_birth'] = df_reduced['year']
person_df['month_of_birth'] = np.nan
person_df['day_of_birth'] = np.nan
person_df.iloc[10]= '...'
person_df.iloc[11]= ' '
person_df.gender_concept_id.iloc[11]= 99.9
person_df.year_of_birth.iloc[11]= 99.9
person_df.person_id.iloc[11]= 'mapped %'
person_df.rename(index={10:'...'}, inplace=True)
#person_df.rename(index={11:'mapped %'}, inplace=True)
person_df.head(12)
person_df.shape

# Step 3: Load the transformed data to the OMOP person table (e.g., database table)
#transformed_data.to_csv('omop_person_table.csv', index=False)
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure(data=[
                     go.Table(
                        header=dict(values=["<b>" + str(n) + "</b>" for n in list(person_df.columns)], 
                                    align=['left','center'],
                                    line_color='darkslategray',
                                    fill_color='lightskyblue',
                                    font=dict(color='black', size=9.5),
                                   ),
                        cells=dict(values=person_df.values.transpose(),
                                   fill_color = [["white","lightgrey"]*person_df.shape[0]],
                                   align='left'
                                  )
                            )
                       ])
#fig.update_layout(width=1000, height=600)
fig.show()
fig.write_image('person_OMOP.png',scale=6)
```
